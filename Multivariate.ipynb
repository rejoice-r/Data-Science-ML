```python
import statsmodels.formula.api as smf
import pandas as pd
from statsmodels.stats.outliers_influence import variance_inflation_factor as vif
```

```python
df = pd.read_csv("https://github.com/OpenClassrooms-Student-Center/Design-Statistical-Models/raw/refs/heads/master/data/advertising.csv")
```

```python
model = smf.ols(formula = 'Sales ~ TV + Radio + Newspaper', data = df)
```

```python
results = model.fit()
print(results.summary())
```

```python
print("Coefficients: {:.5f}".format(results.params['Newspaper']))
print("Coefficients : {:2f}".format(results.pvalues['Newspaper']))
```

```python
m2 = 'Sales ~ TV + Radio'
res2 = smf.ols(m2, data = df).fit()
print(res2.summary())
```

```python
m3 = 'Sales ~ Radio + TV * Newspaper'
res3 = smf.ols(m3, data = df).fit()
print(res3.summary())
```

```python
for num in range(0, 3):
  print(vif(df[['TV', 'Newspaper', 'Radio' ]].values, num))
  num = num + 1
```
