```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from scipy import stats
pd.options.display.float_format = '{:,.2f}'.format
```

```python
df = pd.read_csv('https://github.com/OpenClassrooms-Student-Center/Design-Statistical-Models/raw/refs/heads/master/data/school.csv')
```

```python
G = df[df.sex == 'f'].height.values
```

```python
B = df[df.sex == 'm'].height.values
```

```python
results = stats.ttest_ind(G,B)
```

```python
print(results)
```

```python
g2 = df[df.sex == 'f'].weight.values
```

```python
b2 = df[df.sex == 'm'].weight.values
```

```python
res1 = stats.ttest_ind(g2,b2)
```

```python
print(res1)
```

```python
sns.histplot(G, label ='Girls', kde=True)
sns.histplot(B, label = 'Boys', kde=True)
plt.legend()
plt.title('Histogram of Girls and Boys heights')
plt.xlabel('Height (in cm)')
sns.despine()
```
