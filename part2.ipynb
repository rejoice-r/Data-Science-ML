```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import statsmodels.formula.api as smf
import statsmodels.api as sm
```

```python
df = pd.read_csv('/day.csv')
df = df[['season', 'temp','hum','windspeed','cnt']]
```

```python
formulas = ['cnt ~ temp', 'cnt ~ temp + windspeed + hum']
for formula in formulas:
  print("----"*20)
  print()
  print()
  results = smf.ols(formula, data = df).fit()
  print(results.summary())
```

```python
for seas in range (1, 5):
  print('---'*20)
  print('Season {}'.format(seas))
  for form2 in formulas:
    res2 = smf.ols(form2, data = df[df.season == seas]).fit()
    print("{} \tR^2: {:.2f} \tpvalues: {:.3f} \tpvalue of intercept:{:.3f}".format(form2, res2.rsquared, res2.pvalues.iloc[1], res2.pvalues.iloc[0] ))
```

```python
res3 = smf.ols('cnt ~ temp + windspeed + hum + season', data = df).fit()
print(res3.summary())
```

```python
df['hum2'] = df.hum**2
df['temp2'] = df.temp**2
df['windspeed2'] = df.windspeed**2
```

```python
formula2 =['cnt ~ temp + temp2 + windspeed + hum', 'cnt ~ temp + windspeed + hum + hum2', 'cnt ~ temp + windspeed + windspeed2 + hum']
for formula in formula2:
  print("-----"*20)
  print()
  print()
  res34 = smf.ols(formula, data = df).fit()
  print(res34.summary())
```
