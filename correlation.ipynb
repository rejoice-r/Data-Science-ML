```python
import matplotlib.pyplot as plt
```

```python
df2 = pd.read_csv('https://github.com/OpenClassrooms-Student-Center/Design-Statistical-Models/raw/refs/heads/master/data/anscombes.csv', sep=';')
```

```python
df2.head()
```

```python
df2.groupby(by='dataset').mean()
```

```python
df2.groupby(by='dataset').std()
```

```python
we = list(np.ndenumerate(df2.dataset.unique()))
print(we)
```

```python
fig,ax = plt.subplots(2,2, figsize=(6,6))

for k,name in np.ndenumerate(df2.dataset.unique()):
  cond = df2.dataset == name
  sing = np.polyfit(df2[cond].x,df2[cond].y, 1)
  sing_x = np.linspace(df2[cond].x.min(), df2[cond].x.max(), 3)
  sing_y = sing[0] * sing_x + sing[1]

  plt.subplot(2,2,k[0]+1)
  plt.scatter(df2[cond].x,df2[cond].y)
  plt.plot(sing_x,sing_y, color = 'orange')
```
