```python
import pandas as pd
import statsmodels.formula.api as smf
import numpy as np
```

```python
df = pd.read_csv('https://github.com/OpenClassrooms-Student-Center/Design-Statistical-Models/raw/refs/heads/master/data/credit_default_sampled.csv')
```

```python
df.head()
```

```python
df.loc[df.default == 'No', 'default'] = 0
df.loc[df.default == 'Yes', 'default'] = 1
```

```python
model = smf.logit('default ~ income + balance', data = df)
result = model.fit()
print(result.summary())
```

```python
prob = 1/(1 + np.exp(-result.fittedvalues))
```
