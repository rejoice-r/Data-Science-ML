```python
import pandas as pd
```

```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('https://github.com/OpenClassrooms-Student-Center/Design-Statistical-Models/raw/refs/heads/master/data/auto-mpg.csv')
```

```python
sns.histplot(df.mpg, kde=True)
```

```python
sns.histplot(df.acceleration, kde=True)
```

```python
df['mpg'] = (df.mpg - np.mean(df.mpg))/np.std(df.mpg)
df['acceleration'] = (df.acceleration - np.mean(df.acceleration))/np.std(df.acceleration)
```

```python
df[['mpg', 'acceleration']].describe().loc[['mean', 'std']]
```

```python
from scipy import stats
```

```python
stats.kstest(df.acceleration, 'norm')
```

```python
df['log_mpg'] = np.log(df.mpg + 10)
```

```python
df['log_mpg'] = (df.log_mpg - np.mean(df.log_mpg))/np.std(df.log_mpg)
```

```python
fig, axes = plt.subplots(1, 2, figsize=(12,6))
ax = plt.subplot(1,2,1)
plt.title('mpg')
sns.histplot(df.mpg, ax=ax, kde=True)
sns.despine()
ax = plt.subplot(1,2,2)
plt.title('log_mpg')
sns.histplot(df.log_mpg, ax=ax, kde=True)
sns.despine()
```

```python
stats.kstest(df.log_mpg, 'norm')
```
